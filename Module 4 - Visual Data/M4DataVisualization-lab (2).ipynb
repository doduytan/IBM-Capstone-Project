## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df['ConvertedComp'].plot(kind = 'hist')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df['Age'].plot(kind = 'box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.plot(kind = 'scatter', x = 'Age', y = 'WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

QUERY = """
SELECT *
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5, sizes=(10, 500)).set(title = 'WorkWeekHrs and CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt

QUERY = """
SELECT DatabaseDesireNextYear, count(*) as count 
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC LIMIT 5
"""
df = pd.read_sql_query(QUERY,conn)
df.set_index('DatabaseDesireNextYear', inplace=True) #reset index
df.plot(kind = 'pie', subplots=True, figsize=(20, 6), autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish To Learn')
plt.legend(labels=df.index, loc='upper right')
plt.axis('equal')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE AGE >= 30 AND AGE <=35
"""
df = pd.read_sql_query(QUERY,conn)
df1 = df.groupby('Age').median()
df1.plot(kind = 'area', stacked=True, figsize=(15, 7))
plt.title('stacked chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
plt.legend(loc = 'best')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE AGE >= 45 AND AGE <=60
"""
df = pd.read_sql_query(QUERY,conn)
df_1 = df.groupby('Age').median()
df_1.plot(kind = 'line', figsize=(20, 6))
plt.title('Median of Developers\' Annual Salary for all ages from 45 to 60')
plt.ylabel('ConvertedComp')
plt.xlabel('Age')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) as MainBranch
from master
group by MainBranch
"""

df=pd.read_sql_query(QUERY,conn)

df.plot(kind='barh', figsize=(10,6), color='lightskyblue')

plt.show()
```
